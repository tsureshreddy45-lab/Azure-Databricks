```python
# 1.Dataframe without strict schema

my_data = [
    (100, "John", "Smith"),
    (200, "Anna", "Smith"),
    (300, "Peter", "Jones")
]

columns = ["id", "firstname", "lastname"]

df = spark.createDataFrame(my_data, columns)
display (df)
```

```python
# Create DataFrame with strict schema
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

emp_data = [(100, "John", "Smith"),
            (200, "Anna", "Smith"),
            (300, "Peter", "Jones")
        ]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("firstname", StringType(), True),
    StructField("lastname", StringType(), True)
])

df = spark.createDataFrame(emp_data, schema)
display(df)
```
